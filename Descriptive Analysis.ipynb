## 2018-19 NBA Regular Season Box Score Data

This data comes from https://www.nbastuffer.com/2018-2019-nba-player-stats/ for the 2018-19 regular season.

```python
import pandas as pd

df = pd.read_csv('2018-2019_NBA_Player_Stats.csv')
df
```

The first thing to notice is that the CSV file includes 622 rows, and some players appear more than once. We can confirm this by running `value_counts()` on the `FULL NAME` column.

```python
df['FULL NAME'].value_counts()
```

```python
len(df['FULL NAME'].unique())
```

```python
df[df['FULL NAME'] == 'Jason Smith']
```

We notice that players who switched teams during the season have their stats recorded separately for each team. We may wish to combine these separate entries by using a weighted average using the `GP` column.

```python
df.isnull().sum()
```

```python
df[df['eFG%'].isnull()]
```

```python
temp = df[df['ORTG'].isnull()]
(temp['MPG'] * temp['GP']).describe()
```

There are a number of rows with missing statistics, such as `eFG%`, `TS%`, and `Tor%`. However, these are a result of very small sample sizes - after all, a player can't have an effective field goal percentage if they did not take any shots. In particular, 25 players have no `ORTG`, but this is mostly for players who played less than 15 minutes, and at most 93 minutes. We may wish to remove all rows where the total number of minutes played is below a certain threshold.

```python
df.describe()
```

```python
df['GP'].sum()/(30*41)
```

This suggests that 21 players saw the floor in the average regular season game - a reasonable figure.

```python
df['TEAM'].value_counts()
```

There are significant fluctuations in the number of players per team. Some teams, such as the Memphis Grizzlies, had almost 30 players see minutes played over the course of the season, compared to only 16 players for the San Antonio Spurs. The teams with more players may have played more G-League players on 10-day contracts, or been involved in a larger number of trades.

```python
df.corr()
```

This matrix, although overwhelming, shows us several quick observations. For instance, `MPG` is (obviously) highly correlated with volume metrics such as `PPG`. Interestingly, `USG%` and `eFG%` have a negative correlation.
