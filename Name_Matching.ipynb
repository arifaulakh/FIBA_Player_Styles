```python
import pandas as pd
import collections
df = pd.read_csv("NBA_INTL_FIBA_Clustering.csv", encoding='latin-1', index_col=0)
df['CleanName'] = clean_names = [s.upper() for s in [''.join([c for c in n if c.isalpha() or c==' ']) for n in df.index]]
c = collections.Counter(clean_names)
sum([v>=2 for v in c.values()])
```

```python
g = df.groupby('CleanName')[[c for c in df.columns if 'Cluster' in c]].count()
g[(g > 1).apply(any, axis=1)]
```
