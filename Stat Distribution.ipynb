## Descriptive Analysis - Distribution of Statistics

First, let's load the unmerged dataframe.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Filter_Unmerged.csv').drop(columns = ['Unnamed: 0'])
df.head()
```

Now let's plot histograms showing the distribution of each of the stats.

```python
cols = df.select_dtypes(include='number').columns
f = plt.subplots(8, 4, figsize=(18, 32))
for i in range(len(cols)):
    plt.subplot(8, 4, i+1)
    plt.hist(df[cols[i]], bins=20)
    plt.title(cols[i])
    plt.tight_layout()
plt.show()
```

Most of the stats appear to be skewed binomial distributions. Normalizing the statistics by assigning z-scores may not be appropriate for all stats. Special care must be taken with `FT%`, `2P%` and `3P%` due to the large number of zeros that don't fit in the distribution.
